```
import numpy as np
x = np.linspace(0,10,1000) # x is a 1000 evenly spaced numbers from 0 to 10
y = np.sin(x)
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y)
plt.show()
```

```
x = np.linspace(0,10,1000) # x is a 1000 evenly spaced numbers from 0 to 10
y = np.sin(x)
fig = plt.figure()
plt.plot(x, y)
plt.show()
```

```
x = np.linspace(0, 10, 1000)
plt.plot(x, np.cos(x));
```

```
x = np.linspace(0,10,1000) # 1darray of length 1000
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('function sin(x)')
plt.show();
```

```
x = np.linspace(0,10,1000) # 1darray of length 1000
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show();
```

```
x = np.linspace(0,10,1000) # 1darray of length 1000
plt.plot(x, np.sin(x), color='k')
plt.plot(x, np.cos(x), color='r', linestyle ='--')
plt.show()
```

```
x = np.linspace(0,2,100) 
y1 = x**2
y2 = x
plt.plot(x, y1, color='k')
plt.plot(x, y2, color='b')
plt.show();
```

```
x = np.linspace(0,10,1000) # 1darray of length 1000
plt.plot(x, np.sin(x), 'k:', label='sin(x)')
plt.plot(x, np.cos(x), 'r--', label='cos(x)')
plt.legend()
plt.show();
```

```
x = np.linspace(0,10,1000) # 1darray of length 1000

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
plt.show();
```

```
import pandas as pd
president_df = pd.read_csv('http://sololearn.com/uploads/files/president_heights_party.csv')
plt.scatter(president_df['height'], president_df['age'])
plt.show()
```

```
plt.scatter(president_df['height'], president_df['age'],
   marker='<',
   color='b')
plt.xlabel('height'); 
plt.ylabel('age')
plt.title('U.S. presidents')
plt.show()
```

```
president_df.plot(kind = 'scatter', 
  x = 'height', 
  y = 'age',
  title = 'U.S. presidents')
plt.show();
```

```
president_df.plot(kind = 'scatter', x = 'order', y = 'height', title = 'U.S. president heights')
plt.show();
```

```
president_df['height'].plot(kind='hist',
  title = 'height',
  bins=5)
plt.show()
```

```
plt.hist(president_df['height'], bins=5)
plt.show() # skewed distribution. The distribution isn’t quite symmetric.
```

```
president_df['height'].hist();
```

```
president_df['age'].plot(kind='hist',
  title = 'Age',
  bins=5, color='b')
plt.show()
```

```
president_df['height'].describe()
```

```
plt.style.use('classic')
president_df.boxplot(column='height');
```

```
plt.style.use('classic')
president_df.boxplot(column=['height','age']);
```

```
party_cnt = president_df['party'].value_counts()

plt.style.use('ggplot')
party_cnt.plot(kind ='bar')
plt.show()
```
